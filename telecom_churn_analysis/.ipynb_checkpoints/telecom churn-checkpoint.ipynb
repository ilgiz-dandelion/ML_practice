### Exploratory Analysis 

```python
import pandas as pd
```

```python
df =pd.read_excel('telecom_churn.xlsx')
df
```

```python
df.info()
```

```python
df.head()
```

```python
df['International plan'] = df['International plan'].map({"Yes": 1, "No": 0})
df.head()
```

```python
df['Voice mail plan'] = df['Voice mail plan'].map({"Yes": 1, "No": 0})
df.head()
```

```python
df.pop("State")
df.head()
```

```python
df['Churn'] = df['Churn'].map({True: 1, False: 0})
df.head()
```

```python
df.isnull().sum()
```

```python
y = df.pop("Churn")
```

```python
X = df.copy()
```

### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Algorithm choice

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.get_params()
```

### Model evaluation

```python
clf.fit(X_train, y_train)
```

### Analysis of simulation results

```python
clf.score(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test = list(y_test)
```

```python
trues = 0

for i in range(len(y_test)):
    if y_test[i] == y_pred[i]:
        trues += 1
        
print(trues)
```

```python
len(y_test)
```

```python
print(trues / len(y_test) * 100)
```

### Model improvement

```python
for n in range(10, 101, 10):
    print(f"Evaluating the model with {n} trees...")
    clf = RandomForestClassifier(n_estimators = n).fit(X_train, y_train)
    print(f"Accuracy on test data: {clf.score(X_test, y_test)}")
```

### The final version of the model. Usage for problem solving

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
```

```python
import pickle
```

```python
pickle.dump(clf, open("my_model1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("my_model1.pkl", "rb"))
```
